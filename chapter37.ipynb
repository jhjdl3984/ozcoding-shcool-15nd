##표준모듈

-모듈 : 여러 변수와 함수를 가지고 있는 집합체
- 모듈 사용방법 => import 모듈이름

-표준모듈 : 파이썬에 기본적으로 내장되어 있는 모듈

-math 모듈 : 수학과 관련된 기능을 가지고 O
- math 모듈 내 함수
    
        sin(x) : 사인 값을 구함

        cos(x) : 코사인 값을 구함

        tan(x) : 탄젠트 값을 구함

        ceil(x) : 올림

        floor(x) : 내림

```python
# math 모듈

import math

print(math.sin(5))          # sin() => 함수
print(math.cos(5))          # cos() => 함수
print(math.tan(5))          # tan() => 함수
print(math.ceil(3.6))       # ceil() => 함수
# 함수를(sin(),cos()등)쓰기 위해서 math라는 모듈을 앞에 써야함 => print(math.sin())
```

####from 구문 => from 모듈이름 import 가져오고싶은변수 or 함수

```python
from math import sin, cos, tan, ceil

print(sin(5))
print(cos(5))
print(tan(5))
print(ceil(3.6))
```

```python
# math 모듈의 모든 함수를 가져오는 방법 => import *

from math import *

print(sin(5))
print(cos(5))
print(tan(5))
print(floor(3.6))
# 모든 함수를 불러 오는거라 시간이 오래걸림
```

####as 구문
=> 모듈을 가져오다 보면 모듈이름의 충돌이 발생하는 경우 or 너무 길어 짧게 줄여서 사용하고 싶은 경우에 사용

- as 구문 사용법 => import 모듈이름 as 식별자




```python
# as 구문

import math as m

print(m.sin(5))
print(m.cos(5))
print(m.tan(5))
print(m.floor(3.6))
```

####os 모듈
- 운영체제와 관련된 기능을 가진 모듈
- os.system 명령어의 경우 사용 시 주의

```python
import os

print("현재 운영체제:", os.name)
print("현재 폴더:", os.getcwd())
print("현재 폴더 정보:", os.listdir())

# 폴더 생성
os.mkdir("oz")

# 폴더 삭제
os.rmdir("oz")

# 파일 생성
with open("oz.txt", 'w') as file:
    file.write("수강생 여러분 안녕하세요")
# 파일이름 변경
os.rename('oz.txt', 'oz2.txt')

# 파일 삭제
os.remove("oz2.txt")

# os.system
os.system("ls")    # 맥은 ls
```

####datetime 모듈 => 날짜와 시간 모듈


```python
import datetime

print("오늘 날짜와 시간 출력")
now = datetime.datetime.now()       # 첫번째 datetime => 모듈
                                    # 두번째 datetime => 클래스
print(now.year, "년")
print(now.month, "월")
print(now.day, "일")
print(now.hour, "시")
print(now.minute, "분")
print(now.second, "초")
```

```python
import datetime

now = datetime.datetime.now()

print("다양한 시간 포맷")

output_1 = now.strftime("%Y.%m.%d %H:%M:%S")    # %y => 25
                                                # %Y => 2025
                                                # %h => Oct(월이 영어로)
                                                # %M, %S => 대문자
                                                # => 시간은 다 대문자로 쓴다고 생각 !!
output_2 = f'{now.year}년 {now.month}월 {now.day}일 {now.hour}시 {now.minute}분 {now.second}초'
output_3 = now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초") # "년월일시분초"앞에 *가 없으면 "년월일시분초"가 맨처음에 있는 %Y{}에 중괄호에 다 들어감

print(output_1)
print(output_2)
print(output_3)
```

####time 모듈
- 시간 모듈
- 특정 시간동안 코드를 정지할 수 있는 기능

```python
import time

print("3초 뒤에 어떤일이 일어날까요?")
time.sleep(3)       # 3초동안 멈춤
print("이건 3초 후에 나타납니다")
```

####urllib 모듈

```python
from urllib import request          # request => 요청

target = request.urlopen("https://www.naver.com/")
web_code = target.read()

print(web_code)
# request.urlopen("https://www.naver.com/") => url을 오픈하도록 => 네이버에 요청
# 출력 결과에 나온것 = 네이버홈에서 오른쪽마우스->검사누르면 나오는 내용
```
